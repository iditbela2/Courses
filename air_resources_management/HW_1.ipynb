# Q-3 Total anthropogentic VOC emmissions between 2011-2015 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [np.arange(2011,2016),[0.26872, 0.198375,0.11533,0.177037,0.18847],[0.1691,0.16927,0.16536,0.1723,0.175055]] #year, indestry, transportationin
```

```python
data[0]
```

```python
%config InlineBackend.figure_format = 'retina'
indestry = data[1]
transportation = data[2]
years = data[0]    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(years, indestry, width)
p2 = plt.bar(years, transportation, width,
             bottom=indestry)

plt.ylabel('Average emmissions [T/h]',labelpad=16)
plt.xlabel('Year',labelpad=12)
plt.title('Average antropogenic VOC emmissions in HBA\n between 2011-2015')
plt.xticks(years, ['2011','2012','2013','2014','2015'])
#plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('indestry', 'transportation'))

fig = plt.gcf()
plt.show()
```
